**Columns Description**

**name**: The name of the restaurant or establishment, as listed on Google Maps.

**type**: The general category or type of the restaurant or establishment. In the provided data, the types are "Indian restaurant" and "Restaurant".

*phone*: The phone number of the restaurant or establishment, as listed on Google Maps.

*full_address*: The full address of the restaurant or establishment, as listed on Google Maps.

*street*: The street address of the restaurant or establishment, as listed on Google Maps.

**city**: The city or locality in which the restaurant or establishment is located, as listed on Google Maps.

**latitude**: The latitude coordinate of the restaurant or establishment, as listed on Google Maps.

**longitude**: The longitude coordinate of the restaurant or establishment, as listed on Google Maps.

**rating**: The overall rating of the restaurant or establishment on Google Maps, as a decimal value out of 5.

**working_hours**: The hours of operation for the restaurant or establishment, listed as a dictionary object with keys representing each day of the week and values representing the hours of operation on that day.

**business_status**: The current operational status of the restaurant or establishment, as listed on Google Maps. In the provided data, both establishments are listed as "OPERATIONAL".

**logo**: A URL linking to the restaurant or establishment's logo, as listed on Google Maps.

**verified**: A boolean value indicating whether the restaurant or establishment has been verified by Google.

**location_link**: A URL linking to the Google Maps page for the restaurant or establishment.

```python

# Define a dictionary that maps English day names to Arabic day names
day_names = {
    "Monday": "الاثنين",
    "Tuesday": "الثلاثاء",
    "Wednesday": "الأربعاء",
    "Thursday": "الخميس",
    "Friday": "الجمعة",
    "Saturday": "السبت",
    "Sunday": "الأحد"
}

# Define a function to convert the day names in a string from English to Arabic
def convert_day_names(s):
    for eng, arb in day_names.items():
        s = s.replace(eng, arb)
    return s

# Apply the function to the working_hours column in the merged_df DataFrame
merged_df['working_hours'] = merged_df['working_hours'].apply(convert_day_names)
```

```python
# Load the numpy array into a variable
arabic_working_hours = merged_df['working_hours'].values

# Loop over each string value in the array and replace the non-breaking narrow space
for i in range(len(arabic_working_hours)):
    arabic_working_hours[i] = arabic_working_hours[i].replace('\u202f', ' ')

# Assign the modified array back to the 'working_hours' column in the DataFrame
merged_df['working_hours'] = arabic_working_hours
```

**Feature engineering**

- Total Reviews (New Feature)

```python
# Define a function that extracts the total number of reviews from a string value
def extract_total_reviews(review_score_str):
    if review_score_str == 'not mention':
        return 0
    else:
        review_score_dict = eval(review_score_str)
        return sum(review_score_dict.values())

# Apply the function to the 'reviews_per_score' column and assign the result to a new column
merged_df['total_reviews'] = merged_df['reviews_per_score'].apply(extract_total_reviews)
```

- popularity (New Feature)

```python
# Popularity score = average rating x number of reviews
merged_df['popularity'] = merged_df['rating'] * merged_df['total_reviews']
```


- Business Size (New Feature)

```python
# Define the business size categories based on the total number of reviews
def categorize_business_size(total_reviews):
    if total_reviews <= 50:
        return 'Small'
    elif total_reviews <= 150:
        return 'Medium'
    else:
        return 'Large'

# Create a new column 'business_size' based on the total number of reviews
merged_df['business_size'] = merged_df['total_reviews'].apply(categorize_business_size)
```


- Business Level (New Feature)

```python
merged_df['rating'] = pd.to_numeric(merged_df['rating'])
# Define the business size categories based on the total number of reviews
def business_level(business_rate):
    if business_rate <= 2:
        return 'Poor'
    elif business_rate <=3 :
        return 'Fair'
    elif business_rate <=4 :
        return 'Good'
    else:
        return 'Excellent'

# Create a new column 'business_size' based on the total number of reviews
merged_df['business_level'] = merged_df['rating'].apply(business_level)
```

- Model

```python
# Identify string columns
string_cols = merged_df.columns[merged_df.dtypes == 'object']
# Convert string columns to lowercase
merged_df[string_cols] = merged_df[string_cols].apply(lambda x: x.str.lower())
```

```python
from sklearn.cluster import KMeans
import pandas as pd
df = merged_df
X = df[['latitude', 'longitude']]
df['name'] = df['name'].replace(to_replace=r'\d', value='', regex=True)

# Determine the optimal number of clusters using the elbow method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#  K-means
optimal_num_clusters = 6
kmeans = KMeans(n_clusters=optimal_num_clusters, init='k-means++', max_iter=300, n_init=10, random_state=42)
kmeans.fit(X)
# Assign cluster labels to each restaurant
labels = kmeans.labels_
# Add cluster labels to the DataFrame
df['cluster'] = labels
```

```python
import pickle
# Save the trained KMeans model to a file
with open('unsupervised_model.pkl', 'wb') as file:
    pickle.dump(kmeans, file)
```

```python
from sklearn.metrics import silhouette_score
avg_silhouette_score = silhouette_score(X, labels)
print(f"The average silhouette score is: {avg_silhouette_score:.2f}")
```

```python
from math import radians, sin, cos, sqrt, atan2

def find_closest_restaurant(user_latitude, user_longitude, restaurant_type):
    # Use the clustering algorithm to determine which cluster the user's location belongs to
    user_location = [[user_latitude, user_longitude]]
    user_cluster = kmeans.predict(user_location)[0]

    cluster_restaurants = df[df['cluster'] == user_cluster]
    cluster_restaurants = cluster_restaurants[cluster_restaurants['type'] == restaurant_type]

    # Calculate the distance between the user and each restaurant in the same cluster
    distances = []
    distances_rating = []

    for i, row in cluster_restaurants.iterrows():

        restaurant_latitude = row['latitude']
        restaurant_longitude = row['longitude']
        restaurant_location = (restaurant_latitude, restaurant_longitude)

        # Calculate the great-circle distance between the user and the restaurant
        R = 6371.0 # radius of the Earth in kilometers
        lat1 = radians(user_latitude)
        lon1 = radians(user_longitude)
        lat2 = radians(restaurant_latitude)
        lon2 = radians(restaurant_longitude)
        dlon = lon2 - lon1
        dlat = lat2 - lat1
        a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
        c = 2 * atan2(sqrt(a), sqrt(1 - a))
        distance = R * c

        # Add the restaurant name, location, distance, location link, and rating to the distances and distances_rating lists
        distances.append((row['name'], restaurant_location, distance, row['location_link'], row['rating']))
        distances_rating.append((row['name'], restaurant_location, distance, row['location_link'], row['rating'],row['phone']))

    # Sort the distances list by distance
    distances.sort(key=lambda x: x[2])

    # Sort the distances_rating list by distance and rating
    distances_rating.sort(key=lambda x: (0.7*x[2] + 0.3*(5-x[4]), x[2]))

    closest_restaurants = distances_rating[:4]

    return closest_restaurants
```

```python
from typing import Final
from telegram import Update, Location, ReplyKeyboardMarkup, KeyboardButton
from telegram.ext import Updater, CommandHandler, MessageHandler, Filters, ConversationHandler

print('Starting up bot...')

TOKEN: Final = # write bot token
BOT_USERNAME: Final = #write BOT_USERNAME


LOCATION, RESTAURANT_TYPE = range(2)

def start(update: Update, context):
    # Ask the user to share their location
    update.message.reply_text('Please share your location.', reply_markup=ReplyKeyboardMarkup(
        [[KeyboardButton(text="Share Location 📍", request_location=True)]], resize_keyboard=True, one_time_keyboard=True))

    return LOCATION


def handle_location(update: Update, context):
    # Get the latitude and longitude from the user's location
    location: Location = update.message.location
    latitude = location.latitude
    longitude = location.longitude

    # Store the user's location in the conversation context
    context.user_data['latitude'] = latitude
    context.user_data['longitude'] = longitude

    # Ask the user for the type of restaurant
    update.message.reply_text('What type of restaurant are you looking for?', reply_markup=ReplyKeyboardMarkup(
        [['Cafe ☕', 'Restaurant 🍴']], resize_keyboard=True, one_time_keyboard=True))

    return RESTAURANT_TYPE


def handle_restaurant_type(update: Update, context):
    # Get the user's chosen restaurant type
    restaurant_type = update.message.text.lower()

    # Store the restaurant type in the conversation context
    context.user_data['restaurant_type'] = restaurant_type

    # Determine the correct type for the clustering algorithm
    if restaurant_type == 'cafe ☕':
        clustering_type = 'cafe'
    elif restaurant_type == 'restaurant 🍴':
        clustering_type = 'restaurant'

    # Store the restaurant type in the conversation context
    context.user_data['restaurant_type'] = restaurant_type
         #,
    # Find the closest restaurants of the given type
    latitude = context.user_data.get('latitude')
    longitude = context.user_data.get('longitude')
    closest_restaurants = find_closest_restaurant(latitude, longitude, clustering_type)

    # Format the responses for the closest restaurants
    responses = []
    for i, restaurant in enumerate(closest_restaurants):
        name = restaurant[0]
        distance = restaurant[2]
        location_link = restaurant[3]
        rating = restaurant[4]
        phone = restaurant[5]
        response = f"{i+1}. {get_display(name)}\n- التقييم: {rating}\n- المسافة من موقعك الحالي: {distance:.2f}km\n- للحجز : {phone}\n- الموقع: {location_link}"
        responses.append(response)

    # Combine the responses and send them to the user
    response = "\n\n".join(responses)
    update.message.reply_text(response)

    # End the conversation
    return ConversationHandler.END


def cancel(update: Update, context):
    update.message.reply_text('Conversation cancelled.')
    return ConversationHandler.END


def error(update: Update, context):
    print(f'Update {update} caused error {context.error}')


if __name__ == '__main__':
    updater = Updater(token=TOKEN, use_context=True)
    dispatcher = updater.dispatcher

    conv_handler = ConversationHandler(
        entry_points=[CommandHandler('start', start)],
        states={
            LOCATION: [MessageHandler(Filters.location, handle_location)],
            RESTAURANT_TYPE: [MessageHandler(Filters.text & ~Filters.command, handle_restaurant_type)]
        },
        fallbacks=[CommandHandler('cancel', cancel)]
    )

    # Add the conversation handler to the bot
    dispatcher.add_handler(conv_handler)
    dispatcher.add_error_handler(error)

    # Start the bot
    print('Polling...')
    updater.start_polling()
    updater.idle()
```
